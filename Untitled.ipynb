```python
import numpy as np
import matplotlib.pyplot as plt

# Load the MNIST dataset from OpenML
mnist = fetch_openml("mnist_784")

# Normalize the data
X = mnist["data"] / 255
y = mnist["target"].astype(int)
# Calculate the covariance matrix
cov_mat = np.cov(X, rowvar=False)

# Calculate the eigenvectors and eigenvalues of the covariance matrix
eigen_values, eigen_vectors = np.linalg.eig(cov_mat)

```

## Select number of optimal features automatically.

```python
# Calculate the explained variance ratio for each principal component
explained_variance_ratio = eigen_values / np.sum(eigen_values)

# Calculate the cumulative explained variance ratio
cumulative_explained_variance_ratio = np.cumsum(explained_variance_ratio)

# Plot the explained variance ratio for each principal component
plt.plot(explained_variance_ratio, label="Explained variance ratio")

# Plot the cumulative explained variance ratio
plt.plot(cumulative_explained_variance_ratio, label="Cumulative explained variance ratio")

plt.legend()
plt.show()
```

```python
# Sort the eigenvalues in decreasing order and select the top k eigenvectors
k = 200
eigen_values_idx = np.argsort(eigen_values)[::-1]
eigen_vectors = eigen_vectors[:, eigen_values_idx[:k]]

# Project the data onto the top k eigenvectors to get the reduced data representation
X_reduced = X @ eigen_vectors

print(X_reduced.shape)
```

### if we want to retain 95% of the variance in the data, we can select the number of features such that the cumulative explained variance ratio is greater than or equal to 0.95.

## Visualize extracted features

```python
X_reduced_array= X_reduced.to_numpy()
```

```python

# Plot the reduced data representation along with the target values
for target in np.unique(y):
    target_idx = np.where(y == target)
    plt.scatter(X_reduced_array[target_idx, 0], X_reduced_array[target_idx, 1], label=str(target))

plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)

plt.legend()
plt.show()
```

## Reconstruct the original images using extracted features

```python
# Reconstruct the original data using the reduced data representation and the top k eigenvectors
X_reconstructed = X_reduced @ eigen_vectors.T
```

```python
X_reconstructed=X_reconstructed.to_numpy()
```

## Classifing the features using Random forest classifier  

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.2, random_state=0)

# Train a random forest classifier on the training data
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)

# Evaluate the classifier on the testing data
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print("Accuracy:", acc)
```
